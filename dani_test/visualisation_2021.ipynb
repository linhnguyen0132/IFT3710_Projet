```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_2021_obs = pd.read_csv("C:/Users/danid/OneDrive/Documents/Math&info/projet_apprentissage/GeoLifeCLEF-GLC21/GeoLifeCLEF-GLC21/data/observations/observations_fr_train.csv", sep=";")
data_2021_obs.head(5)
data_2021_obs.columns
```

```python
# Enlever les nans
data_2021_obs = data_2021_obs.dropna()
len(data_2021_obs)
train_obs = data_2021_obs[data_2021_obs["subset"] == "train"]
```

```python
# Visualiser les données avec seaborn 
# Plus spécifiquement nous voulons regarder le nombre d'occurence de chaque espèce
plt.figure(figsize=(6, 4))
sns.histplot(data=train_obs, x="species_id")
plt.tight_layout()
plt.show()

# Interprétation : Nous voyons que les données de France de 2021 présentent un left-skew. 
```
